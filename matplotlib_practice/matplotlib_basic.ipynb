```python
# Histogram
x_hist1 = np.random.normal(20,5,50) # mean 20, median 5 and sample of 50
x_hist1 
```

```python
x_hist1 = np.random.normal(20,5,1000) # mean 20, median 5 and sample of 1000
plt.hist(x_hist1)
plt.show()
```

```python
x_hist1 = np.random.normal(20,5,1000) # mean 20, median 5 and sample of 1000
plt.hist(x_hist1, bins=5)
plt.show()
```

```python
x_hist1 = np.random.normal(20,5,1000) # mean 20, median 5 and sample of 1000
plt.hist(x_hist1, bins=[x_hist1.min(),18,19,20,21, x_hist1.max()])
plt.show()
```

```python
x_hist1 = np.random.normal(20,5,1000) # mean 20, median 5 and sample of 1000
plt.hist(x_hist1,cumulative=True)
plt.show()
```
