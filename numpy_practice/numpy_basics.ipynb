```python
import numpy as np
```

```python
a = np.array([10,20,30,40,50,60,70,80,90])
print(a)
print(type(a))
print(a[3]) # its 40
print(a[1:5])
a[1:5]
print(a[:5])
a[3] = 340 # changing the value
print(a)
```

```python
# multi-dimension array
a_mul = np.array([
    [10,20,30],
    [25,45,55],
    [50,60,70]
]) 
print(a_mul[0])
print(a_mul[0,2]) # 30
```
